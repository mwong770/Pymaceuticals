### Analysis

*This analysis uses data on the physiological responses of mice given one of four drugs (Capomulin, Infubinol, Ketapril, or a placebo) to treat squamous cell carcinoma (SCC), a common form of skin cancer. 250 mice were treated with the drugs for 45 days.*

* Capomulin performed much better than Infubinol, Ketapril, or the placebo at treating SCC. The mean tumor volume decreased by 19% in mice taking Capomulin over the course of treatment, while the mean tumor volume increased in mice treated with Infubinol (46%), Ketapril (57%), or the placebo (51%).

* Mice on Capomulin had less new metastatic sites than mice on other treatments. By the end of the study, on average, mice taking Capomulin had only one new metastatic site. Mice taking Infubinal had roughly two new metastatic sites, and mice taking a placebo or Ketapril had roughly 3 new metastatic sites. 

* Mice treated with Capomulin had the highest rate of survival. 84% of mice taking Capomulin survived the 45 day treament. Mice taking Ketapril had the same rate of survival as those taking the placebo, 44%. Mice taking Infubinol had the lowest rate of survival, 36%.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# City File to Load 
drug_data_to_load = os.path.join('data', 'mouse_drug_data.csv')

drug_df = pd.read_csv(drug_data_to_load)

print(drug_df.shape)
drug_df.head()
```

```python
# File to Load (Remember to Change These)
response_data_to_load = os.path.join('data', 'clinicaltrial_data.csv')

response_df = pd.read_csv(response_data_to_load)

print(response_df.shape)
response_df.head()
```

```python
# Combine the data into a single dataset
drug_response_df = pd.merge(response_df, drug_df, how='left', on='Mouse ID')
drug_response_df.head()
```

```python
# Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_time_gb = drug_response_df.groupby(['Drug', 'Timepoint'])
mean_tumor_vol_s = drug_time_gb['Tumor Volume (mm3)'].mean()
mean_tumor_vol_s

mean_tumor_vol_df = pd.DataFrame(mean_tumor_vol_s).reset_index()


mean_tumor_vol_df.head()
```

```python
# Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_s = drug_time_gb['Tumor Volume (mm3)'].sem()

sem_tumor_df = pd.DataFrame(sem_tumor_s).reset_index()

sem_tumor_df.head()
```

```python
# Pivot Mean Data Frame
mean_tumor_vol_pivot_df = mean_tumor_vol_df.pivot(index='Timepoint', columns='Drug')['Tumor Volume (mm3)']

mean_tumor_vol_pivot_df
```

```python
# Pivot SEM Data Frame
sem_tumor_pivot_df = sem_tumor_df.pivot(index='Timepoint', columns='Drug')['Tumor Volume (mm3)']

sem_tumor_pivot_df
```

```python
# Plot (with Error Bars)
plt.errorbar(mean_tumor_vol_pivot_df.index, 
             mean_tumor_vol_pivot_df['Capomulin'], 
             yerr=sem_tumor_pivot_df['Capomulin'], 
             fmt='ro--', markersize=4, linewidth=.8) 

plt.errorbar(mean_tumor_vol_pivot_df.index, 
             mean_tumor_vol_pivot_df['Infubinol'], 
             yerr=sem_tumor_pivot_df['Infubinol'], 
             fmt='b^--', markersize=4, linewidth=.8)

plt.errorbar(mean_tumor_vol_pivot_df.index, 
             mean_tumor_vol_pivot_df['Ketapril'], 
             yerr=sem_tumor_pivot_df['Ketapril'], 
             fmt='gs--', markersize=4, linewidth=0.8)

plt.errorbar(mean_tumor_vol_pivot_df.index, 
             mean_tumor_vol_pivot_df['Placebo'], 
             yerr=sem_tumor_pivot_df['Placebo'], 
             fmt='kd--', markersize=4, linewidth=0.8)

plt.title('Tumor Response to Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Time (Days)')

plt.grid()

plt.legend(loc='best', fancybox=True, fontsize='small')

plt.show()
```

```python
plt.savefig(os.path.join('images', 'tumor_response.png'))
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site_s = drug_time_gb['Metastatic Sites'].mean()

mean_met_site_df = pd.DataFrame(mean_met_site_s).reset_index()

mean_met_site_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_met_site_s = drug_time_gb['Metastatic Sites'].sem()

sem_met_site_df = pd.DataFrame(sem_met_site_s).reset_index()

sem_met_site_df.head()
```

```python
# Re-Format the Mean Data Frame
mean_met_site_pivot_df = mean_met_site_df.pivot(index='Timepoint', columns='Drug')['Metastatic Sites']

mean_met_site_pivot_df
```

```python
# Re-Format the SEM Data Frame
sem_met_site_pivot_df = sem_met_site_df.pivot(index='Timepoint', columns='Drug')['Metastatic Sites']

sem_met_site_pivot_df
```

```python
# Plot (with Error Bars)
plt.errorbar(mean_met_site_pivot_df.index, 
             mean_met_site_pivot_df['Capomulin'], 
             yerr=sem_met_site_pivot_df['Capomulin'], 
             fmt='ro--', markersize=4, linewidth=.8, capsize=2) 

plt.errorbar(mean_met_site_pivot_df.index, 
             mean_met_site_pivot_df['Infubinol'], 
             yerr=sem_met_site_pivot_df['Infubinol'], 
             fmt='b^--', markersize=4, linewidth=.8, capsize=2)

plt.errorbar(mean_met_site_pivot_df.index, 
             mean_met_site_pivot_df['Ketapril'], 
             yerr=sem_met_site_pivot_df['Ketapril'], 
             fmt='gs--', markersize=4, linewidth=0.8, capsize=2)

plt.errorbar(mean_met_site_pivot_df.index, 
             mean_met_site_pivot_df['Placebo'], 
             yerr=sem_met_site_pivot_df['Placebo'], 
             fmt='kd--', markersize=4, linewidth=0.8, capsize=2)

plt.title('Metastatic Spread During Treatment')
plt.ylabel('Met. Sites')
plt.xlabel('Treatment Duration (Days)')

plt.grid()

plt.legend(loc='best', fancybox=True, fontsize='small')

plt.show()
```

```python
plt.savefig(os.path.join('images', 'metastatic_sites.png'))
```

## Survival Rates

```python
# Count of Mice Grouped by Drug and Timepoint 
survival_ct_s = drug_time_gb['Mouse ID'].count()

survival_ct_df = pd.DataFrame(survival_ct_s).rename(columns={'Mouse ID': 'Mouse Count'}).reset_index()

survival_ct_df.head()
```

```python
# Re-Format the Count Data Frame
survival_ct_pivot_df = survival_ct_df.pivot(index='Timepoint', columns='Drug')['Mouse Count']

survival_ct_pivot_df
```

```python
# Plot Survival Percentage
mouse_start_ct = 25

plt.plot(survival_ct_pivot_df['Capomulin']*100/mouse_start_ct, 'ro--', markersize=4, linewidth=0.8)
plt.plot(survival_ct_pivot_df['Infubinol']*100/mouse_start_ct, 'b^--', markersize=4, linewidth=0.8)
plt.plot(survival_ct_pivot_df['Ketapril']*100/mouse_start_ct, 'gs--', markersize=4, linewidth=0.8)
plt.plot(survival_ct_pivot_df['Placebo']*100/mouse_start_ct, 'kd--', markersize=4, linewidth=0.8)
         
plt.title('Survival During Treatment')
plt.ylabel('Survival Rate (%)')
plt.xlabel('Time (Days)')

plt.grid()

plt.legend(loc='best', fancybox=True)

plt.show()
```

```python
plt.savefig(os.path.join('images', 'survival_rates.png'))
```

## Summary Bar Graph

```python
# Mean Percent Changes for Each Drug
tumor_per_change_s = (mean_tumor_vol_pivot_df.iloc[-1] - mean_tumor_vol_pivot_df.iloc[0])*100 / mean_tumor_vol_pivot_df.iloc[0]
tumor_per_change_s
```

```python
# Relevant Percent Changes 
tumor_per_changes_subgp = (tumor_per_change_s['Capomulin'], 
               tumor_per_change_s['Infubinol'], 
               tumor_per_change_s['Ketapril'], 
               tumor_per_change_s['Placebo'])
tumor_per_changes_subgp
```

```python
# Splice the data between passing and failing drugs
failing = [x for x in tumor_per_changes_subgp  if x > 0]
passing = [x for x in tumor_per_changes_subgp if x <= 0 ]

failing_idx = [index for index, x in enumerate(tumor_per_changes_subgp) if x > 0]
passing_idx = [index for index, x in enumerate(tumor_per_changes_subgp) if x <= 0 ]

print(f'failing: {failing}')
print(f'passing: {passing}')

print(f'failing index: {failing_idx}')
print(f'passing index: {passing_idx}')
```

```python
# Create Chart
passing_drugs = plt.bar(passing_idx, passing, width=1, color='g')
failing_drugs= plt.bar(failing_idx, failing, width=1, color='r')

# Configure Chart
tick_locations = np.arange(len(tumor_per_changes_subgp)) + 0.5

plt.title('Tumor Change Over 45 Day Treatment', fontsize=15)
plt.ylabel('% Tumor Volume Change', fontsize=12)
plt.xticks(tick_locations,['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'])

plt.ylim([-30,70])
plt.grid(linestyle='dashed')

# Label the Percentages of Change
def set_text_labels(drugs):
    for bar in drugs:
        value = bar.get_height()
        formatted_value = '{:.0f}%'.format(value)
        x_location = bar.get_x() + 0.5
        if value > 0:
            plt.text(x_location, 1, formatted_value, color='w', fontsize=15, ha='center', va='bottom')
        else:
            plt.text(x_location, -3, formatted_value, color='w', fontsize=15, ha='center', va='top')
        
set_text_labels(passing_drugs)
set_text_labels(failing_drugs)

plt.savefig(os.path.join('images', 'passing_failing_barchart.png'))

plt.show()
```
